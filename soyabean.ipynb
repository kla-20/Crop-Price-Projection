### Evaluation Metric function

```python
# Evaluation Metrics

def Emetrics(y_test,y_pred):
  from sklearn import metrics

  # Print result of MAE
  print(f"Mean Absoulute error is : {metrics.mean_absolute_error(y_test,y_pred)}")

  # Print result of MSE
  print(f"Mean Squared error is: {metrics.mean_squared_error(y_test,y_pred)}")

  # Print result of RMSE
  print(f"Root Mean Square error is: {np.sqrt(metrics.mean_squared_error(y_test,y_pred))}")

  # Print result of R2
  print(f"R2 score is: {metrics.r2_score(y_test,y_pred)}")
```

### Backward Stepwise selection

```python
X = df_action.loc[:, z_score_features]
y = df_action.loc[:, target]
```

```python
import statsmodels.api as sm

def backward_selection(X, y, significance_level=0.05):
    remaining_features = list(X.columns)
    selected_features = list(X.columns)
    while len(remaining_features) > 0:
        X_train = X[selected_features]
        model = sm.OLS(y, sm.add_constant(X_train)).fit()
        p_values = model.pvalues.iloc[1:]
        max_p_value = p_values.max()
        if max_p_value > significance_level:
            feature_to_remove = p_values.idxmax()
            selected_features.remove(feature_to_remove)
        else:
            break
    return selected_features

selected_features = backward_selection(X, y)
```
